```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Başlıklar ve yazılar listeleri
yazilar = []
basliklar = []

for j in range(1, 5):
    page = requests.get('https://www.karar.com/yazarlar/zafer-acar?page=' + str(j))
    soup = BeautifulSoup(page.content, 'html.parser')
    linkler = soup.find_all('article', {'class': 'item box-shadow'})
    for link in linkler:
        try:
            link1 = 'https://www.karar.com' + link.a.get('href')
            baslik = link.h3.text.strip()  # Başlık olarak h3 elemanının metnini al
            page2 = requests.get(link1)
            soup2 = BeautifulSoup(page2.content, 'html.parser')

            makale = soup2.select('article.article-content')[0].text.strip()  # Makale içeriğini al

            basliklar.append(baslik)  # Başlıklar listesine ekle
            yazilar.append(makale)  # Yazılar listesine ekle

        except (AttributeError, IndexError, requests.exceptions.RequestException) as e:
            print(f"Hata: {e}")
            continue

# DataFrame oluşturma
df = pd.DataFrame({
    'Title': basliklar,
    'Text': yazilar
})

# DataFrame'i görüntüle
print(df)
```

# NlP Preorocessing

```python
# Text Preorocessing 
def preprocess_text(df):
    df['title_text'] = df['title_text'].str.lower()  # Tüm metni küçük harfe dönüştürür
    df['title_text'] = df['title_text'].str.replace('[^\w\s]', '', regex=True)  # Noktalama işaretlerini kaldırır
    df['title_text'] = df['title_text'].str.replace('\d+', '', regex=True)  # Rakamları kaldırır
    df['title_text'] = df['title_text'].str.replace('\n', ' ', regex=True)  # Satır sonlarını boşlukla değiştirir
    df['title_text'] = df['title_text'].str.replace('\r', '', regex=True)  # Carriage return karakterlerini kaldırır
    df['title_text'] = df['title_text'].str.replace('\s+', ' ', regex=True)  # Fazla boşlukları tek boşlukla değiştirir
    df['title_text'] = df['title_text'].str.strip()  # Başta ve sondaki boşlukları kaldırır
    return df

df = preprocess_text(df)
```

# Tokenization (Tokenlere ayırma )
### Tokenizasyon, metni kelimelere, cümlelere veya diğer anlamlı parçalara ayırma işlemidir.

```python
def token_text(text):
    return word_tokenize(text)
```

# Stop Words Kaldırma (Gereksiz kelimeleri kaldırma)

```python
def remove_stop_words(tokens):
    stop_words = set(stopwords.words('turkish'))
    return [word for word in tokens if word not in stop_words]
```

# Lemmatizasyon (Lemmatization) (Kök Bulma)  
####  Stematizasyon da kullanılabilir  ama bir tane kullanmak yeterlidir  ve daha hızlı  .
#### Lemmatizasyon kelimeleri anlamlı köklerine indirgerken, stemleme sadece kelimenin bir kısmını alır. Genelde sadece lemmatizasyon kullanılır çünkü anlamı korur.

```python
def lemmatize_tokens(tokens):
    lemmatizer = WordNetLemmatizer()
    return [lemmatizer.lemmatize(word) for word in tokens]
```

```python
df['Tokens'] = df['title_text'].apply(token_text)
df['Tokens_No_Stop_Words'] = df['Tokens'].apply(remove_stop_words)
df['Lemmatized_Tokens'] = df['Tokens_No_Stop_Words'].apply(lemmatize_tokens)
```

```python
stop_words = {'ilk', 'bir', 'ancak', 'sonra', 've', 'ya', 'bile'}
def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)
df['no_stopwords'] = df.apply(lambda row: remove_stopwords(row['Title']) + ' ' + remove_stopwords(row['Text']), axis=1)
```

# Sentiment Analysis

```python
from textblob import TextBlob

def get_sentiment(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity

    if polarity > 0.20:
        sentiment = 'pozitif'
    elif polarity < -0.20:
        sentiment = 'negatif'
    else:
        sentiment = 'natural'

    polar = polarity
    return pd.Series([sentiment, polar])

df[['sentiment', 'polarity']] = df['title_text'].apply(lambda x: get_sentiment(x))
```
